```python
# Programa para determinar cuál número es mayor

numero1 = int(input("Ingresa el primer número: "))
numero2 = int(input("Ingresa el segundo número: "))

if numero1 > numero2:
    print("El número mayor es:", numero1)

elif numero2 > numero1:
    print("El número mayor es:", numero2)

else:
    print("Ambos números son iguales.")
```

```python
# Programa para clasificar una edad

edad = int(input("Ingresa tu edad: "))

if edad < 18:
    print("Menor de edad")

elif edad <= 59:
    print("Adulto")

else:
    print("Adulto mayor")
```

```python
# Programa para determinar si un año es bisiesto

anio = int(input("Ingresa un año: "))

if (anio % 4 == 0 and anio % 100 != 0) or (anio % 400 == 0):
    print("El año es bisiesto")

else:
    print("El año no es bisiesto")
```

```python
# Programa para evaluar una calificación

calificacion = float(input("Ingresa una calificación entre 0 y 10: "))

if calificacion < 0 or calificacion > 10:
    print("Calificación no válida")

elif calificacion >= 9.1:
    print("Excelente")

elif calificacion >= 8.1:
    print("Notable")

elif calificacion >= 7.0:
    print("Aprobado")

else:
    print("Reprobado")
```

```python
# Programa para calcular descuentos

compra = float(input("Ingresa el monto de la compra: "))

if compra > 500:
    descuento = compra * 0.15

elif compra >= 300:
    descuento = compra * 0.10

else:
    descuento = 0

total = compra - descuento

print("Descuento aplicado: $", descuento)
print("Total a pagar: $", total)
```
